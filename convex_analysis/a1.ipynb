```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the convex function f(x) = x^2 and its derivative
def f(x):
    return x**2

def df(x):
    return 2 * x

# Tangent line at a point x0
def tangent_line(x0, x):
    return f(x0) + df(x0) * (x - x0)

# Set up the plot
x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

# Define the point where we draw the tangent line
x0 = 1  # You can change this to move the tangent point
tangent_vals = tangent_line(x0, x_vals)
```

```python

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the convex function f(x)
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2$', color='blue')

# Plot the tangent line at x0
plt.plot(x_vals, tangent_vals, label=f'Tangent at x0={x0}', linestyle='--', color='orange')

# Highlight the point of tangency
plt.scatter([x0], [f(x0)], color='red', zorder=5)
plt.text(x0, f(x0) + 0.5, f'Point of tangency at x0={x0}', color='red')

# Add labels and legend
plt.title('Convex Function and Tangent Line')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Parameters of the ellipse (major and minor axes)
a = 2  # Semi-major axis
b = 1  # Semi-minor axis

# Parametric equation of the ellipse: (x(t), y(t)) for t in [0, 2pi]
def ellipse(t):
    x = a * np.cos(t)
    y = b * np.sin(t)
    return x, y

# Derivatives for the tangent line
def ellipse_derivative(t):
    dx_dt = -a * np.sin(t)
    dy_dt = b * np.cos(t)
    return dx_dt, dy_dt

# Tangent line at a given parameter t0
def tangent_line(t0, t):
    x0, y0 = ellipse(t0)
    dx_dt, dy_dt = ellipse_derivative(t0)
    
    # Parametric equation of the tangent line (linear approximation)
    x_tangent = x0 + dx_dt * (t - t0)
    y_tangent = y0 + dy_dt * (t - t0)
    return x_tangent, y_tangent

# Plot settings
t_vals = np.linspace(0, 2 * np.pi, 400)  # Parametric range for ellipse
x_vals, y_vals = ellipse(t_vals)

# Tangent at a specific point on the ellipse (let's choose t0 = pi/4)
t0 = np.pi / 4
x0, y0 = ellipse(t0)
x_tangent, y_tangent = tangent_line(t0, t_vals)
```

```python
# Create the plot
plt.figure(figsize=(8, 6))

# Plot the ellipse
plt.plot(x_vals, y_vals, label='Ellipse', color='blue')

# Plot the tangent line at the point (x0, y0)
plt.plot(x_tangent, y_tangent, label=f'Tangent at t0={t0:.2f}', linestyle='--', color='orange')

# Highlight the point of tangency
plt.scatter([x0], [y0], color='red', zorder=5)
plt.text(x0, y0 + 0.1, f'Point of tangency at t0={t0:.2f}', color='red')

# Add labels and legend
plt.title('Ellipse and Tangent Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Display the plot
plt.grid(True)
plt.gca().set_aspect('equal')  # Set equal scaling for x and y axes
plt.show()
```
